```python
#Loading data and preparing data for classification
```

```python
x_train = np.load('F:\Kaggle_Competetion\\Final data\\t_data.npy')
x_train.shape
```

```python
y_df = pd.read_csv('F:\Kaggle_Competetion\\Final data\\y_data.csv')
print(type(y_df))
```

```python
y_train = y_df.to_numpy()
print(type(y_train))
```

```python
y_train = y_train.reshape(-1,)
y_train
```

```python
#Visualizing Data
```

```python
classes = ["0_No_DR","1_Mild","2_Moderate","3_Severe","4_Proliferative_DR"]
```

```python
def plot_sample(x,y,index):
    plt.imshow(x[index])
    plt.xlabel(classes[y_train[index]])
```

```python
plot_sample(x_train,y_train,27566)
```

```python
#Normalizing Values between 0 & 1
x_train = x_train/255
```

```python
#Defining CNN model
#CNN
cnn = models.Sequential([
    layers.Conv2D(filters=16, kernel_size=(6, 6), activation='relu', input_shape=(100, 80, 3)),
    layers.MaxPooling2D((2, 2)),
    
    layers.Conv2D(filters=32, kernel_size=(6, 6), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

cnn.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
cnn.fit(x_train, y_train, epochs=5)
```
